```python
from matplotlib import pyplot as plt
def plot_block_ram_cpu(x, y1,y2,name):
    plt.title(name)

    plt.plot(
        x,
        y1,
        label = 'CPU usage',
        linestyle="-."
    )
    plt.plot(
        x,
        y2,
        label = 'RAM usage',
        linestyle="--"
    )
    plt.legend()
    plt.show()
  
def plot_block(x,y,name):
    plt.title(name)
    plt.xlabel('EspaÃ§o (Mb)', color= 'black')
    plt.ylabel('Tempo (s)', color= 'black')
    plt.plot(
        x,
        y,
    )   
```
